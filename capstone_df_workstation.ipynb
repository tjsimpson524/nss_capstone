```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../nss_capstone/df.csv').drop(columns = 'Unnamed: 0')
```

```python
df['Total Difference'] = df['Open Total'] - df['Close Total']
df['Spread Difference'] = df['Open Spread'] - df['Close Spread']
```

```python
df.replace(0.0, np.nan, inplace=True)
```

```python
df['Open Spread'] = df.groupby('game_id')['Open Spread'].transform(lambda x: x.ffill().bfill().iloc[0] if x.notnull().any() else np.nan)
df['Close Spread'] = df.groupby('game_id')['Close Spread'].transform(lambda x: x.ffill().bfill().iloc[0] if x.notnull().any() else np.nan)
```

```python
df['Open Total'].fillna(df.groupby('game_id')['Open Total'].transform('max'), inplace=True)
df['Close Total'].fillna(df.groupby('game_id')['Close Total'].transform('max'), inplace=True)
```

```python
df.loc[df['ML'] > 0, ['Open Spread', 'Close Spread']] = '+' + df[['Open Spread', 'Close Spread']].astype(str)
```
