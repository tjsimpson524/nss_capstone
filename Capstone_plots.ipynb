```python
import plotly.graph_objects as go
from ipywidgets import interact

def create_histogram(ML_pct):
    filtered_df = df[df['ML_pct'] == ML_pct]
    game_counts = filtered_df.shape[0] // 2
    
    win_counts = filtered_df[filtered_df['W/L_bool'] == 1].shape[0] // 2
    loss_counts = filtered_df[filtered_df['W/L_bool'] == 0].shape[0] // 2
    
    win_ratio = win_counts / game_counts
    loss_ratio = loss_counts / game_counts

    labels = ['Favorite Wins', 'Underdog Wins']
    counts = [win_counts, loss_counts]
    
    colors = 'lightblue'
    
    fig = go.Figure()
    fig.add_trace(go.Bar(x=labels, y=counts, marker=dict(color=colors)))
    fig.update_layout(
        title=f'Win Ratios Based on Closing Moneyline Implied Probability % = {ML_pct}',
        xaxis=dict(title='Wins'),
        yaxis=dict(title='Games')
    )
    fig.add_annotation(
        x=labels[0],
        y=max(counts),
        text=f'{win_ratio:.2f}',
        showarrow=False,
        font=dict(size=12)
    )
    fig.add_annotation(
        x=labels[1],
        y=max(counts),
        text=f'{loss_ratio:.2f}',
        showarrow=False,
        font=dict(size=12)
    )
    fig.show()
    
    fig.write_html("efficiency_plot.html")

interact(create_histogram, ML_pct=(0.5, 0.99, 0.01))
```

```python
def create_histogram(ML_pct):
    filtered_df = df[df['ML_pct'] == ML_pct]
    game_counts = filtered_df.shape[0] // 2  # Divide the number of rows by 2 to get the count of games
    
    win_counts = filtered_df[filtered_df['W/L_bool'] == 1].shape[0] // 2  # Divide the count by 2 for games
    loss_counts = filtered_df[filtered_df['W/L_bool'] == 0].shape[0] // 2  # Divide the count by 2 for games
    
    win_ratio = win_counts / game_counts
    loss_ratio = loss_counts / game_counts

    labels = ['Favorite Wins', 'Underdog Wins']
    counts = [win_counts, loss_counts]
    
    colors = ['palegoldenrod', 'darkslategrey']
    
    fig = px.bar(x=labels, y=counts, color=labels, color_discrete_sequence=colors)
    fig.update_layout(
        title=f'Win Ratios Based on Closing Moneyline Implied Probability % = {ML_pct}',
        xaxis_title='W/L_bool',
        yaxis_title='Count'
    )
    fig.add_annotation(
        x=0, y=max(counts), text=f'{win_ratio:.2f}', showarrow=False, font=dict(size=12)
    )
    fig.add_annotation(
        x=1, y=max(counts), text=f'{loss_ratio:.2f}', showarrow=False, font=dict(size=12)
    )
    
    # Save the interactive plot as an HTML file
    pyo.plot(fig, filename='../nss_capstone/efficiency_plot.html')

```
